```python
# List of characters to remove
unwanted_chars = '<=>£¦«°»éü˜–—‘’“”•…`'

# Create a translation table that maps each unwanted character to None
translation_table = str.maketrans('', '', unwanted_chars)

# Specify file names (make sure the original dataset file is in the same directory)
input_file = 'dataset.txt'
output_file = 'cleaned_dataset.txt'

# Read the original dataset
with open(input_file, 'r', encoding='utf-8') as f:
    text = f.read()

print("Original number of characters:", len(text))

# Remove the unwanted characters
cleaned_text = text.translate(translation_table)
print("Cleaned number of characters:", len(cleaned_text))

# Write the cleaned dataset to a new file
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(cleaned_text)

print(f"Cleaned dataset saved to '{output_file}'.")

# Provide a download link for the cleaned dataset (works in Jupyter Notebook)
from IPython.display import FileLink, display
display(FileLink(output_file))
```
